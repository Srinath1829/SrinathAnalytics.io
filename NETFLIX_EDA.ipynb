```python
# importing Libraries


import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns 
```

```python
# File path with showing  'show_ID' as index_col 

netflix_data = pd.read_csv(r"C:\Users\91967\OneDrive\Desktop\code.basic\netflix_titles_2021.csv" , index_col= 'show_id')
```

```python
# Showing data set of netflix shows 

netflix_data
```

```python
# data types of data set and columns 

netflix_data.info()
```

```python

# Data Cleaning and Handling Missing Values

 # Remove ' min' from duration
netflix_data['duration'] = netflix_data['duration'].str.replace(' min', '') 


# Remove ' Seasons' from duration
netflix_data['duration'] = netflix_data['duration'].str.replace(' Seasons', '')


 # Remove ' Season' from duration
netflix_data['duration'] = netflix_data['duration'].str.replace(' Season', '') 

netflix_data
```

```python
# convertions of data set for statistical analysis 

# # Convert duration to numeric
netflix_data['duration'] = pd.to_numeric(netflix_data['duration'], errors='coerce') 


# Fill NaN with mean duration
netflix_data['duration'].fillna(netflix_data['duration'].mean(), inplace=True)  

netflix_data
```

```python
# Replacing 'NaN' values to 'unknown'

# Fill NaN in director column
netflix_data['director'].fillna('Unknown', inplace=True) 

# Fill NaN in cast column
netflix_data['cast'].fillna('Unknown', inplace=True)  


# Fill NaN in country column
netflix_data['country'].fillna('Unknown', inplace=True)  

netflix_data
```

```python
# Data visualization of data set 

#distrubution of show types 


plt.figure(figsize=(8, 6))
sns.countplot(data=netflix_data, x='type', palette='pastel')
plt.title('Distribution of Show Types')
plt.xlabel('Show Type')
plt.ylabel('Number of Shows')
plt.show()
```

```python
#  Insight:

# Netflix has a larger collection of Movies compared to TV Shows,
#  indicating that Movies are more prevalent on the platform.
```

```python
#distribution of show durations with minutes 

plt.figure(figsize=(10, 6))
sns.histplot(data=netflix_data, x='duration', bins=30, color='skyblue')
plt.title('Distribution of Show Durations')
plt.xlabel('Duration (minutes)')
plt.ylabel('Number of Shows')
plt.show()
```

```python
# Insight:

# Most shows have durations between 40 to 120 minutes.
# Longer duration shows are relatively rare, suggesting that viewers prefer content of moderate length.

```

```python
# top 10 countries 

top_countries = netflix_data['country'].value_counts().head(10)


#Top 10 countries by show production by no.of shows


plt.figure(figsize=(12, 8))
sns.barplot(x=top_countries.values, y=top_countries.index, palette='muted')
plt.title('Top 10 Countries by Show Production')
plt.xlabel('Number of Shows')
plt.ylabel('Country')
plt.show()
```

```python
# Insight:

# The United States is the top contributor to Netflix content, followed by India and the United Kingdom.
# This showcases the platform's global diversity in content.
```

```python
# Distribution of ratings 

plt.figure(figsize=(8, 6))
sns.countplot(data=netflix_data, x='rating', palette='dark')
plt.title('Distribution of Ratings')
plt.xlabel('Rating')
plt.ylabel('Number of Shows')
plt.xticks(rotation=45)
plt.show()
```

```python
# Insight:

# The majority of shows are rated as 'TV-MA' (Mature Audience) and 'TV-14' (Teens 14 or older), 
# suggesting that the platform offers content catering to different age groups.
```

```python
# Group data by release_year and count the number of shows/movies
yearly_counts = netflix_data['release_year'].value_counts().sort_index()

# Plot time series
plt.figure(figsize=(10, 6))
sns.lineplot(x=yearly_counts.index, y=yearly_counts.values, marker='o', color='b')
plt.title('Yearly Trend of Shows/Movies Added')
plt.xlabel('Release Year')
plt.ylabel('Number of Shows/Movies')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
# Release year vs duration 

plt.figure(figsize=(10, 6))

sns.regplot(data=netflix_data, x='release_year', y='duration', scatter_kws={'alpha':0.3}, color='purple')
plt.title('Release Year vs. Duration')
plt.xlabel('Release Year')
plt.ylabel('Duration (minutes)')
plt.show()
```

```python
# Insights: 

1. The plot shows show duration against release year.
2. No strong linear trend between release year and duration.
3. Recent years show a slight upward trend in duration.
4. Outliers with longer durations exist across years.
5. Duration changes influenced by various factors.

#This insight highlights that show durations haven't significantly changed with release years,
#but outliers and other factors impact show lengths.
```

```python
# Select specific genres for the heatmap
selected_genres = ['Dramas', 'Comedies', 'Documentaries', 'Action & Adventure']

# Filter the data for the selected genres
filtered_data = netflix_data[netflix_data['listed_in'].isin(selected_genres)]

# Count occurrences by genre and release year
heatmap_data = filtered_data.groupby(['listed_in', 'release_year']).size().unstack(fill_value=0)

# Create a heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(data=heatmap_data, cmap='YlGnBu', annot=True, fmt='g')
plt.title('Genre and Release Year Heat Map')
plt.xlabel('Release Year')
plt.ylabel('Genre')
plt.show()
```

```python
# insights:

The heatmap showcases the distribution of selected genres ('Dramas', 'Comedies', 'Documentaries', 'Action & Adventure') across different release years.

1. Darker colors indicate higher show counts, while lighter colors represent lower counts.

2.Dramas: Consistently popular with growth in recent years.

3.Comedies: Maintains a consistent presence, slight decline recently.

4.Documentaries: Steady growth, significant rise in recent years.

5.Action & Adventure: Fluctuating popularity with peaks and troughs.

6.Peaks in color intensity denote years of higher genre-specific show counts.

The heatmap informs Netflix's content strategy, highlighting trends and diversification efforts.

# Helps identify regional and global genre preferences.

# Enables strategic planning by aligning with viewer trends
```
